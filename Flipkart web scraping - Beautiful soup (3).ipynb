# Importing important Libraries for web scraping

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
import re 
import pandas as pd
import numpy as np

```

```python
url = urlopen("https://www.flipkart.com/q/best-laptops-under-rs-50000")
flipkart_laptop_data = url.read() # rquest for open url
#print(flipkart_laptop_data)
```

# Convert data into Beautifulsoup data

```python
soupdata_flipkart = soup(flipkart_laptop_data)
```

```python
#dir(soupdata_flipkart) #findAll
```

```python
containers = soupdata_flipkart.findAll('div',{'class':'_2kHMtA'})
print(type(containers),len(containers))
```

```python
f = open("E:/Flipkart Web Scraping Project.csv",'wb')
f.write('Product_name,Ratings,Reviews,stars,Current_Price,MRP,Processor,RAM,Storage,Image_URL\n'.encode())
for container in containers:
    # Finding product name.
    Product = container.findAll('div',{'class':'_4rR01T'})
    Product_name = Product[0].text.split('-')[0].strip().replace(',','')
        
    # Finding Rating & reviews of product
    Rating= container.find('span',{'class':'_2_R_DZ'})
   
    try:
        rating_reviews = re.findall('\d+,?\d*',Rating.text)
        Ratings = rating_reviews[0].replace(',','')
        Reviews = rating_reviews[1].replace(',','')
        
    except:
        Ratings = 0
        Reviews = 0
    
    
    
    
    # Finding stars
    stars = container.find('div',{'class':'_3LWZlK'})
    stars = stars
    try:
        stars = stars.text.replace(',','')
    except:
        stars = 0
        
    #Finding current price.
    Current_Price = container.find('div',{'class':'_30jeq3 _1_WHN1'}).text.replace(',','')
    
    # Finding MRP.
    MRP = container.find('div',{'class':'_3I9_wc _27UcVY'})
    try:
        MRP =MRP.text.replace(',','')
    except:
        print(0)
        
    # Finding Spectifications of product
    Specifications = container.findAll('li',{'class':'rgWa7D'})
    Processor = Specifications[0].text.replace(',','')
    RAM = Specifications[1].text.replace(',','')
    Storage = Specifications[3].text.replace(',','')
    
    
    Image_URL = container.img
    Image_URL = Image_URL.get('src')
    
    print(Product_name,Ratings,Reviews,stars,Current_Price,MRP,Processor,RAM,Storage,Image_URL)
    print('\n')
    f.write(f"{Product_name},{Ratings},{Reviews},{stars},{Current_Price},{MRP},{Processor},{RAM},{Storage},{Image_URL}\n".encode())
    print('\n')
f.close()
```
